```python
import numpy as np
import pandas as pd
import math
```

```python
dfResults = pd.read_csv('results.csv')
```

###### Number of reads which don't map on reference genome

```python
numberOfNonMatched = 0

for index, row in dfResults.iterrows():
    if(math.isnan(row['Custom 0 -3 -7'])):
        numberOfNonMatched += 1

print("Total number of reads", dfResults.shape[0])
print("Number of no matched reads is", numberOfNonMatched)
```

##### Histrogram: x-axis bins with alignment score, y-axis number of reads in group with that score

```python
import matplotlib.pyplot as plt

for col in dfResults.columns[3:]:
    print(col)
    x = dfResults[col].dropna().tolist()
    
    plt.figure(figsize=[10,8])

    plt.hist(x, bins = 10)
    plt.xlabel('Range of alignment score', fontsize=15)
    plt.ylabel('Number of reads in range', fontsize=15)
    plt.xticks(fontsize=15)
    plt.yticks(fontsize=15)
    plt.style.use('ggplot')
    plt.title('Histogram '+ col,fontsize=15)
    
    plt.savefig("Histograms/Histogram_" + col + ".png")
    plt.show()
```
